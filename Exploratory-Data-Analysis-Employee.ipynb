## outliers years at company
- years at company di cek karena untuk mengetahui kenapa beberapa pegawai bisa bertahan lama,setelah di cek outliers yang ada di years at company tidak mengganggu. karena faktor dari kolom lain. yaitu age, dilihat dari age yang sudah tua, pastinya years at company nya juga lebih lama dari yang lain

```python
Q1 = df.describe()['YearsAtCompany']['25%']
Q3 = df.describe()['YearsAtCompany']['75%']
IQR = Q3 - Q1
Upper_Fence = Q3 + (1.5 * IQR)
Lower_Fence = Q1 - (1.5 * IQR)
print(IQR)
```

```python
df[(df['YearsAtCompany'] < Lower_Fence) | (df['YearsAtCompany'] > Upper_Fence)]
```

```python
plt.hist(df['YearsAtCompany'])
```

## outliers Performance Rating
- untuk outliers performance rating itu terdapat pada nilai 4 , 4 itu masih termasuk kategori nilai. maka tidak mengganggu

```python
Q1 = df.describe()['PerformanceRating']['25%']
Q3 = df.describe()['PerformanceRating']['75%']
IQR = Q3 - Q1
Upper_Fence = Q3 + (1.5 * IQR)
Lower_Fence = Q1 - (1.5 * IQR)
print(IQR)
df[(df['PerformanceRating'] < Lower_Fence) | (df['PerformanceRating'] > Upper_Fence)]
```

```python
sns.boxplot(df['PerformanceRating'],palette='vlag')
```
